*  discrete probability distribution

*  tells how many times an event is likely to occur over a specified period.

* It is a count distribution, the parameter of which is lambda (Î»); the mean number of events in the specific interval.

*  the variable can only take whole number values (0, 1, 2, 3, etc.), with no fractions or decimals.

* Many economic and financial data appear as count variables, such as how many times a person becomes unemployed in a given year

For example, if the average number of people who buy cheeseburgers from a fast-food chain on a Friday night at a single restaurant location is 200, a Poisson distribution can answer questions such as, "What is the probability that more than 300 people will buy burgers?

if a video store averaged 400 customers every Friday night, what would have been the probability that 600 customers would come in on any given Friday night?

If the mean is very large, then the Poisson distribution is approximately a normal distribution.



### When should the Poisson Distribution be used?

when the variable in question is a count variable. For instance, to estimate how many defective products will come off an assembly line given different inputs.

## How to generate a Poisson Distribution

```python
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
from scipy.stats import poisson
```

```python
sample1 = np.random.poisson(3,1000)
sample2 = np.random.poisson(10,1000)
sample3 = np.random.poisson(17,1000)
```

```python
sns.kdeplot(sample1,fill=True,label='lambda=3')
sns.kdeplot(sample2,fill=True,label='lambda=10')
sns.kdeplot(sample3,fill=True,label='lambda=17')
plt.title('Poisson Distribution')
plt.show()
```

Examples of random variables which can be represented using a Poisson Distribution are: number of visitors a website gets in a minute, number of customers entering a supermarket in a day, number of cars passing by a street camera in a second.
